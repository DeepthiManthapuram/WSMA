```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag, ne_chunk
```

```python
nltk.download('punkt_tab', quiet = True)
nltk.download('averaged_perceptron_tagger_eng', quiet = True)
nltk.download('stopwords', quiet = True)
nltk.download('wordnet', quiet = True)
nltk.download('omw-1.4', quiet = True)
```

```python
text = """Ram studies at standford. He enjoys coding in python and AI."""
```

```python
#tokenize
sentence = sent_tokenize(text)
words = word_tokenize(text)
words
```

```python
#POS tagging
pos_tags = pos_tag(words)
pos_tags
```

```python
#stopword removal
filtered = [w for w in words if w.lower() not in stopwords.words('english') and w.isalpha()]
filtered
```

```python
#stemming
stemmed = [PorterStemmer().stem(w) for w in filtered]
stemmed
```

```python
#Lemmatization
lemmatized = [WordNetLemmatizer().lemmatize(w) for w in filtered]
lemmatized
```

```python
print("Sentences: ", sentence)
print("Words: ", words)
print("POS tags: ", pos_tags)
print("Filtered: ", filtered)
print("Stemmed: ", stemmed)
print("Lemmatized: ", lemmatized)
```
